## part1 and part2

```python
newDf = pd.DataFrame() 
newDf['CGPA'] = df['CGPA']
newDf['Chance of Admit'] = df['Chance of Admit']
newDf
```

```python
yTrue = newDf['Chance of Admit']
xTrue = newDf['CGPA']
m = len(yTrue)
```

```python
xMean = xTrue.mean()
yMean = yTrue.mean()
xMinusMean = xTrue - xMean
yMinusMean = yTrue - yMean
xy = np.multiply(xMinusMean, yMinusMean)
x2 = np.multiply(xMinusMean, xMinusMean)
tetta1 = np.sum(xy) / np.sum(x2)
tetta0 = yMean - tetta1*xMean
```

```python
def hypothesisFunction(x) :
    ans = tetta0 + tetta1*x
    return ans
def costFunction(m, x) :
    yPredict = hypothesisFunction(x)
    MSE = np.square(np.subtract(yTrue,yPredict)).mean()/2
    return MSE
```

```python
ax1 = newDf.plot.scatter(x = "CGPA", y = "Chance of Admit")
yExp = hypothesisFunction(xTrue)
plt.plot(xTrue, yExp, "#00cc00")
```

```python
costFunction(m, xTrue)
```
